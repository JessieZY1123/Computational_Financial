### Question 1 (Problem 13.5 of the textbook modified)

### A stock price is currently 100. Over each of the next two six-month periods it is expected to go up by 10% (i.e. by a multiple factor of u = 1.1) or down by a multiple factor of d where d = 1/u. The risk-free interest rate is 8% per annum with continuous compounding. What is the value of a one-year European call option with a strike price of 100 using Binomial Model? 

**Type your code below**

```python
S0 = 100
u = 1.1
d = 1/u
K = 100
r = 0.08
period = 2
T = 1.0
dt = T / period
sigma = log(u)/sqrt(dt)
model = BinomialModel(pricing_date, r)
symbol ='Problem 13.5'
stock = Stock(symbol, S0, sigma)
option = Option(Option.Type.CALL,Option.Style.EUROPEAN,stock,T,K)
call = EuropeanCallOption(stock, T, K)
call_price = model.calc_model_price(call, period)
print("Call price is",call_price)
```

### Question 2. Using the same assumption from Question 1. Calculate the European Put option with the same strike and expiry using Binomial Tree. Calculate also implied put price from the call price using Put-Call parity and the implied call price from the put price. Does Put-Call Parity hold for Binomial trees?

```python
option_put = Option(Option.Type.PUT,Option.Style.EUROPEAN,stock,T,K)
put = EuropeanPutOption(stock,T,K)
put_price = model.calc_model_price(put, period)
print('Put price is ',put_price)

corresponding_option_call = model.calc_parity_price(put, put_price)
corresponding_option_put = model.calc_parity_price(call, call_price)
print("corresponding call price is {} when put price is {}".format(corresponding_option_call,put_price))
print("corresponding put  price is {} when call price is {}:".format(corresponding_option_put,call_price))

print("Put-Call Parity holds for Binomial trees ")
```

### Question 7 (Problem 19.3 of the textbook)

### Calculate the delta of an at-the-money six-month European call option on a non-dividend-paying stock when the risk-free interest rate is 10% per annum and the stock price volatility is 25% per annum. 

**Type your code here**

```python
S_Q7= 100
K7=100
T7=6/12
r7=0.1
sigma7=0.25
symbol7 = 'Problem 19.3'
model7 = BlackScholesModel(pricing_date, r7)
stock7 = Stock(symbol7, S_Q7, sigma7)
option_7_call = Option(Option.Type.CALL,Option.Style.EUROPEAN,stock7,T7,K7)
delta7= model7.calc_delta(option_7_call)
print("Delta is",delta7)
```

### Question 8. Calculate the gamma, theta, vega and rho of the same European call option in Question 7

**Type your code here**

```python
gamma = model7.calc_gamma(option_7_call)
theta = model7.calc_theta(option_7_call)
vega = model7.calc_vega(option_7_call)
rho = model7.calc_rho(option_7_call)
print("Gamma is",gamma)
print("Theta is",theta )
print("Vega is",vega)
print("Rho is",rho)
```

### Question 9: (Optional for undergraduate) Plot the graph of Delta, Gamma, Theta, Vega and Rho for the Call Option described in Question 7

```python
x = np.array([i for i in range(50,150)])
delta7 = []
gamma7 = []
theta7 = []
vega7 = []
rho7 = []
for i in x:
    stock7 = Stock(symbol7, i, sigma7)
    option_7_call = Option(
        Option.Type.CALL,
        Option.Style.EUROPEAN,
        stock7,
        T7,
        K7
    )
    delta7.append(model7.calc_delta(option_7_call))
    gamma7.append(model7.calc_gamma(option_7_call))
    theta7.append(model7.calc_theta(option_7_call))
    vega7.append(model7.calc_vega(option_7_call))
    rho7.append(model7.calc_rho(option_7_call))
     
y_delta = np.array(delta7)
y_gamma = np.array(gamma7)
y_theta = np.array(theta7)
y_vega = np.array(vega7)
y_rho = np.array(rho7)

plt.plot(x, y_theta,'g-')
plt.plot(x, y_vega,'r-')
plt.plot(x, y_rho,'b-')
plt.xlabel('stock price')
plt.ylabel('factors')
plt.title('Sensitivities to Call Option')
plt.text(50, 40, 'green: theta', fontsize = 10)
plt.text(50, 35, 'red: vega', fontsize = 10)
plt.text(50, 30, 'blue: rho', fontsize = 10)
plt.show()
```
